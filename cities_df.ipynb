```python
#import dependancies
import pandas as pd
```

```python
# Set the column width

pd.options.display.max_colwidth = 200

# Read in the dataset. 

cities_df = pd.read_csv("./Resources/worldcities.csv")
cities_df
```

```python
# transform data
# drop rows where latitude and longitude are null, since we won't be able to use those values.

cities_df_drop1 = cities_df.dropna(subset=['lat'])
cities_df_drop1
```

```python
cities_df_drop2 = cities_df_drop1.dropna(subset=['lon'])
```

```python
cities_df_dropped3 = cities_df_drop2.dropna(subset=['capital'])
cities_df_dropped3
```

```python
# fill any null values in the csv with 'NA'.

cities_df_dropped4 = cities_df_dropped3.fillna("NA")
cities_df_dropped5= cities_df_dropped4.fillna("NA")
cities_df_dropped5.info()
```

```python
# drop 'capital' column, as that is the column with the largest amount of null values.
# Capital column is also not important for our queries. 

column_to_drop = 'capital'
cities_df_dropped6 = cities_df_dropped5.drop(column_to_drop, axis=1)
cities_df_dropped6
```

```python
# drop 'admin_name' column, as the accent marks may mess with later queries. 

column_to_drop = 'admin_name'
cities_df_dropped7 = cities_df_dropped6.drop(column_to_drop, axis=1)
```

```python
# replace original 'city' column with 'city_ascii'
# since 'city' has accent marks and may cause an issue later down the line. 

column_to_drop = 'city'
cities_df_dropped8 = cities_df_dropped7.drop(column_to_drop, axis=1)


cities_df_dropped8
```

```python
cities_df_dropped8.rename(columns={'lat': 'city'})

cities_df_dropped8.head()
```

```python
# check our dataset for any null valuess
# ensure all the rows are the same length

cities_df = cities_df_dropped8
```

```python
# ensure city was replaced smoothly.

cities_df.rename(columns={'city_ascii': 'city'})
```

```python
# save our dataframe as a CSV for future use. 
cities_df.to_csv('citiesDF.csv')
```
